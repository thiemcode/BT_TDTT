```python
#W5A1
a,b= map(int,input().split())
def Max_Of_Two(a,b):
    if a > b:
        return a
    elif a < b:
        return b
    else:
        return 'so khong hop le'
print(Max_Of_Two(a,b))
```

```python
#W5A3
def prime(n):
    if n < 2:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
n = int(input())
if prime(n):
    print(n,'la so nguyen to')
else:
    print(n,'khong phai so nguyen to')
```

```python
#W5A4
n = int(input())
def sohoanhao(n):
    if n <= 2:
        return False
    s = 0
    for i in range(1,n):
        if n % i == 0:
            s+=i
    return s==n
if sohoanhao(n):
    print(n,'la so hoan hao')
else:
    print(n,'khong phai so hoan hao')
```

```python
#W5A6
n = int(input())
def giaithua(n):
    if n > 0:
        s = 1
        for i in range(1,n+1):
            s *= i
        return s
    return False
if giaithua(n):
    print('giai thua:',giaithua(n))
else:
    print('so khong hop le')
```

```python
#W5A7
n1,n2 = map(int,input().split())
pheptoan = input()
def operat(n1,n2):
    if pheptoan == '+': return n1+n2
    if pheptoan == '-': return n1-n2
    if pheptoan == '*': return n1*n2
    if pheptoan == '/': return n1/n2
print(f"{operat(n1,n2):.2f}")
```

```python
#W5A8
def hamming(n,m):
    so1=[]; so2=[]
    while n!=0:
        so1.append(n%2); n//=2
    while m!=0:
        so2.append(m%2); m//=2
    so1=so1[::-1]; so2=so2[::-1]
    L=max(len(so1),len(so2))
    so1=[0]*(L-len(so1))+so1
    so2=[0]*(L-len(so2))+so2
    dem=0
    for i in range(L):
        if so1[i]!=so2[i]: dem+=1
    return dem
n,m = map(int,input().split())
print(hamming(n,m))
```

```python
#W5A9
n = int(input())
def tong(n):
    if n<=0: 
        return False
    s=0
    while n!=0:
        s += n%10
        n//=10
    return s
if tong(n):
    print('tong cac chu so cua so nguyen duong n:', tong(n))
else:
    print('so khong hop le')
```

```python
#W5A10
n = list(input())
m = list(input())
def dangcau(n,m):
    for x,y in zip(n,m):
        if n.index(x) != m.index(y):
            return False
    return True
print(dangcau(n,m))
    
```
